## 1 The Turing Test:

“The Turing Test” refers to a proposal made by Turing (1950) as a way of dealing with the question whether machines can think, or rather think intelligently like a human.

Turing describes the following kind of game hence it is called "the imitation game". Suppose that we have a person, a machine, and an interrogator. The interrogator is in a room that is separated from the other person and the machine. The objective would be for the interrogator to determine which of the other two is the person, and which is the machine. The interrogator knows the other person and the machine by the labels ‘X’ and ‘Y’—but, at least at the beginning of the game, does not know which of the other person and the machine is ‘X’—and at the end of the game says either ‘X is the person and Y is the machine’ or ‘X is the machine and Y is the person’. The interrogator is allowed to put questions to the person and the machine of the following kind: “Will X please tell me whether X plays chess?” Whichever of the machine and the other person is X must answer questions that are addressed to X. The object of the machine is to try to cause the interrogator to mistakenly conclude that the machine is the other person; the object of the other person is to try to help the interrogator to correctly identify the machine. - Taken from the source (1)

## 2 History of Chatbots:

__ELIZA:__ It was considered to be the first chatbot in the history of Computer Science. ELIZA was developed by Joseph Weizenbaum at Massachusetts Institute of Technology (MIT). It was in 1994 that the term ‘Chatterbot” was coined. ELIZA operates by recognising key words or phrases from the input and proceeds to reproduce a response from its pre-programmed responses according to the keywords. For instance, if a human says that ‘My mother cooks good food’. ELIZA would pick up the word ‘mother’, and respond by asking an open- ended question ‘Tell me more about your family’. This created an illusion of understanding and simulates an interaction with a real human being though the process was a mechanised one.

__ALICE:__ It was developed in 1995 by Richard Wallace. ALICE chatbot was able to use natural language processing (NLP), which allowed for more sophisticated conversation. It was revolutionary, though, for being open-source. Developers could use AIML (artificial intelligence markup language) to create their own chatbots powered by ALICE.

__JABBERWACKY:__ Jabberwacky is a chatterbot created by British programmer Rollo Carpenter. Its stated aim is to “simulate natural human chat in an interesting, entertaining and humorous manner”. It is an early attempt at creating an artificial intelligence through human interaction. The stated purpose was to create an artificial intelligence that would be capable of passing the Turing Test. It is designed to mimic human interaction and to carry out conversations with users. It is not designed to carry out any other functions. Recent developments do allow a more scripted, controlled approach to sit atop the general conversational AI, aiming to bring together the best of both approaches, and usage in the fields of sales and marketing is underway. The ultimate intention is that the program move from a text based system to be wholly voice operated-learning directly from sound and other sensory inputs. JABBERWACKY's creator believes that it can be incorporated into objects around the home such as robots or talking pets, intending both to be useful and entertaining, keeping people company.

__MITSUKU:__ Mitsuku is a chatbot created from AIMLtechnology by Steve Worswick. It claims to be an 18-year-old female chatbot from Leeds, England. It contains all of Alice’s AIML files, with many additions from user generated conversations, and is always a work in progress. Her intelligence includes the ability to reason with specific objects. For example, if someone asks “Can you eat a house?”, Mitsuku looks up the properties for “house”. Finds the value of “made_from” is set to “brick” and replies “no”, as a house is not edible. She can play games and do magic tricks at the user’s request. In 2015 she conversed, on average, in excess of a quarter of a million times daily.

## 4.5 Format the data:

The dialog data should be parsed into question and reponse pairs and this can be done using regular expressions. Further, at this point, the content of the data is not important but rather it is the match up between the response to the question.

```python
# User define Func:

def questions_ans_pairs(data):
    convo_pairs = re.findall(": (.*?)(?:$|\n)", data)
    convo_list.extend(list(zip(convo_pairs, convo_pairs[1:])))
```

```python
# initialise with empty list:
convo_list = []

# Map the user define function onto the data:
convo.map(questions_ans_pairs);

# Convert 'convo' to Dataframe:
convo_df = pd.Series(dict(convo_list)).to_frame().reset_index()
convo_df.columns = ['q', 'a']
```

## 5 Match up the closest question to the its related user input:

## 5.1 Import the required packages:

Import the tf-idf vectorization library and the cosine similarity library.

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.metrics.pairwise import cosine_similarity
```

## 5.2 Match up:

To do this, the training data has to be converted into a tf-idf matrix, where it can then be analysed in terms of matching up the user's input to that in the training set by using cosine similarity.

Cosine Similarity info. link: https://www.machinelearningplus.com/nlp/cosine-similarity/

```python
# Save the model as an object:
vectoriser = TfidfVectorizer(ngram_range=(1,3))

# Apply the vectoriser on the 'question' data:
vectors = vectoriser.fit_transform(convo_df['q'])
```

```python
# Set my custom question:
my_custom_question = vectoriser.transform(['Hi. My name is John.'])

# Calc. the Similarity:
cos_sim = cosine_similarity(my_custom_question, vectors)

# Out the response:
question_response = pd.Series(cos_sim[0]).sort_values(ascending = False)

# Get the top 5 responses:
top5_question_response = question_response.iloc[0:5]

# Inspect:
top5_question_response
```

## 5.3 Test a number of statements:

To do this, a function would be needed to test any number of statements put forward to the machine and return a response.

```python
# User Defined Fucntion:
def get_response(q):
    
    my_question = vectoriser.transform([q])
    cos_sim = cosine_similarity(my_question, vectors)
    question_response = pd.Series(cos_sim[0]).sort_values(ascending = False)
    response_idx = question_response.index[0]
    
    return convo_df.iloc[response_idx]['a']
```

## 6.2 The Problem of Long-Term Dependencies:

RNNs are able to predict the next oncoming or rather the present word that would complete the phrase in question. There are cases where the gap betweem the relevant information and the placement that it's needed is small, this means that the RNNs would be able to learn to use past information to do this. There are other cases where the gap may become very large, as the phrase in question is very long and to get the relevant context, the model have to learn information from further back. In these cases, RNNs may become unable to learn to connect the information as that gap grows.

## 6.2 Long Short Term Memory:

LSTMs also have this chain like structure as RNNs, but the repeating module itself have a different structure. Instead of having a single neural network layer, there are four, interacting in a very special way. This is shown in the diagram below.

```python
# LSTM: Image 1 
display(Image(image_path+'LSTM Image 1.png', width=600, unconfined=True))
print('Image ref -> https://colah.github.io/posts/2015-08-Understanding-LSTMs/')
```

The LSTM model have the ability to removeadd information to the cell state which are carefully regulated by structures called gates. These gates are a way to optionally let information through and are composed of a sigmoid neural net layer along a pointwise multiplication operation. The sigmoid layer outputs are between zero and one, where a value of zero means nothing will go through while a value of one means let everything through the network. An LSTM has three of these gates, to protect and control the cell state.

### 6.2.1 LSTM Walkthrough - Step 1:

This step of the LSTM model was designed to decide what information to be removed from the cell state, which is made by the sigmoid layer that is called the "forget gate". 


```python
# LSTM: Step 1
display(Image(image_path+'LSTM Step 1.png', width=600, unconfined=True))
print('Image ref -> https://colah.github.io/posts/2015-08-Understanding-LSTMs/')
```

### 6.2.2 LSTM Walkthrough - Step 2:

This step is to decide what new information are going to stored in the cell state. There are two parts here. First, a sigmoid layer called the “input gate layer” decides which values to update. Next, a tanh layer creates a vector of new candidate values that could be added to the state.

```python
# LSTM: Step 2
display(Image(image_path+'LSTM Step 2.png', width=600, unconfined=True))
print('Image ref -> https://colah.github.io/posts/2015-08-Understanding-LSTMs/')
```

### 6.2.3 LSTM Walkthrough - Step 3:

This step combines these two gates to create an update to the state. For example, it is now time to update the old cell state, Ct−1, into the new cell state Ct. The previous steps already decided what to do, whereas this step just need to implement it.

```python
# LSTM: Step 3
display(Image(image_path+'LSTM Step 3.png', width=600, unconfined=True))
print('Image ref -> https://colah.github.io/posts/2015-08-Understanding-LSTMs/')
```

### 6.2.4 LSTM Walkthrough - Step 4:

Finally, the last step is to decide what output would be. This output will be based on the filtered version of the cell state. First, run a sigmoid layer which decides what parts of the cell state is going to be outputed. Then, put the cell state through tanh (to push the values to be between −1 and 1) and multiply it by the output of the sigmoid gate, so that it only output the parts that were the ones decided.

```python
# LSTM: Step 4
display(Image(image_path+'LSTM Step 4.png', width=600, unconfined=True))
print('Image ref -> https://colah.github.io/posts/2015-08-Understanding-LSTMs/')
```

## 7 Building the LSTM Model:

## 7.1 Import the required Packages:

```python
# For AMD GPU and have PlaidML installed:
# os.environ["KERAS_BACKEND"] = "plaidml.keras.backend"
```

```python
from keras.models import Model
from keras.layers import Input, LSTM, Dense

import numpy as np
```

## 7.2 Setup the Training Parameters:

```python
# Batch size for training:
batch_size = 64

# Number of epochs for training:
epochs = 100

# Latent dimensionality of the Encoding Space:
latent_dim = 256

# Number of samples to train on:
num_samples = 1000
```

## 7.3 Data Processing:

This section will take the data to be reformatted and vectorised. Note: This model will actually work by generating one character at a time.

```python
# Create lists for the questions and answers:
input_texts = []
target_texts = []

# Create Sets of individual characters for the questions and answers:
input_characters = set()
target_characters = set()
```

```python
# Limit the question and answer paris to 50 characters or less. To speed up the training process:
convo_df['q len'] = convo_df['q'].astype('str').apply(lambda x: len(x))

convo_df['a len'] = convo_df['a'].astype('str').apply(lambda x: len(x))

convo_df = convo_df[ (convo_df['q len'] < 50) & (convo_df['a len'] < 50) ]
```

```python
# Set up the input and target text lists:
input_texts = list(convo_df['q'].astype('str'))

# add a tab and a newline character to the target text.
# This will be the start and stop tokens for the decoder later on.
target_texts = list(convo_df['a'].map(lambda x: '\t' + x + '\n').astype('str'))
```

```python
# Save the distinct characters from the text data into sets:
for input_text, target_text in zip(input_texts, target_texts):
    # Input Characters:
    for char in input_text:
        if char not in input_characters:
            input_characters.add(char)
            
    # Target Characters:
    for char in target_text:
        if char not in target_characters:
            target_characters.add(char)
```

#### Additional Data Preparation:

For the data to work correctly with the model, it will require padding up to the maximum length of the data. Even though any length of the data can be fed into the model, and it will return any length.

```python
# Maximum Padding:  

# Sort the characters of the words:
input_characters = sorted(list(input_characters))
target_characters = sorted(list(target_characters))

# Encoder and Decoder tokens:
num_encoder_tokens = len(input_characters)
num_decoder_tokens = len(target_characters)

# Maximum encoder and decoder length:
max_encoder_seq_len = max( [len(txt) for txt in input_texts] )
max_decoder_seq_len = max( [len(txt) for txt in target_texts] )
```

```python
# 
print('Number of Samples:', len(input_texts))
print('Number of Unique input Tokens:', num_encoder_tokens)
print('Number of Unique output Tokens:', num_decoder_tokens)
print('Max sequence length for inputs:', max_encoder_seq_len)
print('Max sequence length for outputs:', max_decoder_seq_len)
```

#### Vectorise the text data:

```python
# Token Index:
input_token_idx = dict( [(char, i) for i, char in enumerate(input_characters)] )
target_token_idx = dict( [(char, i) for i, char in enumerate(target_characters)] )

# Initialise arrays of zeros:
encoder_input_data = np.zeros( shape = (len(input_texts), max_encoder_seq_len, num_encoder_tokens), dtype='float32' )

decoder_input_data = np.zeros( shape = (len(input_texts), max_decoder_seq_len, num_decoder_tokens), dtype='float32' )

decoder_target_data = np.zeros( shape = (len(input_texts), max_decoder_seq_len, num_decoder_tokens), dtype='float32' )
```

```python
# Inspect:
target_token_idx
```

```python
# One-hot-Encode:
for i, (input_text, target_text) in enumerate(zip(input_texts, target_texts)):
    for t, char in enumerate(input_text):
        encoder_input_data[i, t, input_token_idx[char]] = 1.
    
    for t, char in enumerate(target_text):
        # decoder_target_data is ahead of decoder_input_data by one timestep.
        decoder_input_data[i, t, target_token_idx[char]] = 1.
        
        if t > 0:
            # decoder_target_data will be ahead by one timestamp and will not include the start character.
            decoder_target_data[i, t - 1, target_token_idx[char]] = 1.
```

## 7.4 Setup the seq-to-seq model: for the encoder and decoder LSTMs.

This section will build the model and utilises some of the example codes that can be found from, https://blog.keras.io/a-ten-minute-introduction-to-sequence-to-sequence-learning-in-keras.html/.

#### Define the input sequence and process it:

```python
encoder_inputs = Input(shape= (None, num_encoder_tokens))

encoder = LSTM(latent_dim, return_state = True)
encoder_outputs, state_h, state_c = encoder(encoder_inputs)
```

#### Discard the "encoder_outputs" and only keep the states:

```python
encoder_states = [state_h, state_c]
```

#### Setup the Decoder by using "encoder_states" as the initial state:

```python
decoder_inputs = Input(shape = (None, num_decoder_tokens))
```

#### Setup the Decoder outputs:

The decoder should return full output sequences, and to return internal states as well. Do not use the return states in the training model, but proceed to use them in inference.

```python
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(decoder_inputs, initial_state=encoder_states)
decoder_dense = Dense(num_decoder_tokens, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)
```

## 7.5 Train on the LSTM model:

This section will define the model by using the encoder and decoder input, and the decoder output. The model is then compiled, fitted to the training text data and finaly will be saved as a .h5 file.

Additionally, the model will be trained on 1,000 data samples that would be split into 80% training and 20% validation sets. The model will also run for 100 epochs.

NOTE: Process may take a while.

```python
# Define the model that will turn "encoder_input_data" & "decoder_input_data" into "decoder_target_data"
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer='rmsprop', loss='categorical_crossentropy')
model.fit([encoder_input_data, decoder_input_data], 
          decoder_target_data,
          batch_size=batch_size,
          epochs=epochs,
          validation_split=0.2)

# Save the model:
model.save('s2s.h5')
```
